```python
#define your own Vehicle class here
#the order of the four required positional arguments should be: make,model,year,mileage
#please use the conventional way to define function names and refer to the following test to adjust function names 
#no need to handle exceptions here

class Vehicle:
    def __init__(self, vmak, vmod, vyr, vml):
        self.__make = vmak
        self.__model = vmod
        self.__year = vyr
        self.__mileage = vml
        
        '''Accessors'''
    
    def get_make(self):
        return self.__make
    def get_model(self):
        return self.__model
    def get_year(self):
        return self.__year
    def get_mileage(self):
        return self.__mileage
    
    '''Mutators'''

    def set_make(self, vmak):
        self.__make = vmak
    def set_model(self, vmod):
        self.__model = vmod
    def set_year(self, vyr):
        self.__year = vyr
    def set_mileage(self, vml):
        self.__mileage = vml
    
    def __str__(self):
        cardata = 'Make: {}; '.format(self.__make)
        cardata += 'Model: {}; '.format(self.__model)
        cardata += 'Year of Manufacture: {}; '.format(self.__year)
        cardata += 'Mileage: {}'.format(self.__mileage)
        return cardata

```

```python
#define your own Full Time Employee class here
#Full Time Employee class inherits from Employee class
#the order of the four required positional arguments should be: name,address,vehicle object,salary
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class FullTimeEmployee(Employee):
    def __init__(self,emp_name, emp_address, v, emp_sal):
        Employee.__init__(self,emp_name, emp_address, v)
        self.__salary = emp_sal
        
    '''Accessor'''
    def get_salary(self):
        return self.__salary
    
    '''Mutator'''
    def set_salary(self,emp_sal):
        self.__salary = emp_sal
    
    def compute_compensation(self):
        self.__taxes = 0.00
        
        if(self.__salary < 45000.00):
            self.__taxes = (self.__salary*0.18)
        elif(self.__salary >= 45000.00 and self.__salary<= 82000.00):
            self.__diff1 = (self.__salary - 45000)
            self.__taxes = ((8100.00)+(self.__diff1*0.28))
        else:
            self.__diff1 = (self.__salary - 82000.00)
            self.__taxes = ((8100.00)+(self.__diff1*.33)+(10360.00))
        self.__compensation = (self.__salary - self.__taxes)
        #print("Tax: {}".format(self.__taxes))
        return ((self.__compensation)/52)
    
    def compute_reimbursement(self, annual_expense):
        if annual_expense <= 10000:
            self.__reimbursement = annual_expense
        if annual_expense > 10000:
            self.__reimbursement = 10000 + ((annual_expense-10000)*0.5)
        return ((self.__reimbursement)/52)
        
    def __str__(self):
        fteStr = super().__str__()
        fteStr = "\nDetails of this Full Time Employee are:" + fteStr
        fteStr += '\nSalary: ${:0.2f}'.format(self.__salary)
        return fteStr
```

```python
#define your own Hourly Employee class here
#Hourly Employee class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,hourly rate
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class HourlyEmployee(Employee):
    def __init__(self,emp_name, emp_address, v, hrswkd, hrrate):
        super().__init__(emp_name, emp_address, v)
        self.__hours_wrkd = hrswkd
        self.__hourly_rate = hrrate
        
    '''Accessor'''
    def get_hours_worked(self):
        return self.__hours_wrkd
    
    def get_hourly_rate(self):
        return self.__hourly_rate
    
    '''Mutator'''
    def set_hours_worked(self, hrswkd):
        self.__hours_wrkd = hrswkd
    
    def set_hourly_rate(self, hrrate):
        self.__hourly_rate = hrrate
        
    def compute_compensation(self):
        if(self.__hours_wrkd <= 40):
            self.__compensation = (self.__hours_wrkd * self.__hourly_rate)
        else:
            self.__diff3 = (self.__hours_wrkd - 40)
            self.__compensation = (40.00 * self.__hourly_rate) + ( self.__diff3 * self.__hourly_rate * 1.80)
        return (self.__compensation)
    
    def compute_reimbursement(self, weekly_expense):
        if weekly_expense <= 100:
            self.__reimbursement = weekly_expense
        if weekly_expense > 100:
            self.__reimbursement = 100
        return ((self.__reimbursement))
        
    def __str__(self):
        hreStr = super().__str__()
        hreStr = "\nDetails of this Hourly Employee are:" + hreStr
        hreStr += '\nHours Worked: {}; Hourly Rate: {}'.format(self.__hours_wrkd,self.__hourly_rate)
        return hreStr
```

```python
#define your own Management class here
#Management class inherits from both Full Time Employee class and Consultant class
#the order of the six required positional arguments should be: name, address,vehicle object, salary, hours worked, project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class Management(FullTimeEmployee,Consultant): 
    def __init__(self,emp_name, emp_address, v, emp_sal, hrswkd, pType):
        #Employee.__init__(self, emp_name, emp_address, v)
        FullTimeEmployee.__init__(self,emp_name, emp_address, v, emp_sal)
        Consultant.__init__(self,emp_name, emp_address, v, hrswkd, pType)  
    
    def compute_compensation(self):
        self.__compensation = FullTimeEmployee.compute_compensation(self) + Consultant.compute_compensation(self)
        return (self.__compensation)
    
    def compute_reimbursement(self, annExp, weekExp):
        self.__reimbursement = FullTimeEmployee.compute_reimbursement(self,annExp)+Consultant.compute_reimbursement(self,weekExp)
        return ((self.__reimbursement))
    
    def __str__(self):
        mStr = super().__str__()
        mStr = "\nDetails of this Management are:" +mStr
        mStr = "\nEmployee Name: " + Employee.get_emp_name(self) + "; Employee Address: " + Employee.get_emp_addr(self)
        mStr +=  "\n" + str(Employee.get_vehicle(self))
        mStr += '\nSalary: ${:0.2f}'.format(FullTimeEmployee.get_salary(self))
        mStr += '\nHours Worked: {}; Project Type: {}'.format(Consultant.get_hours_worked(self),Consultant.get_project_type(self))
        return mStr
```

```python
#complete this get_vehicle_input() function to prepare for the main menu application
#this function is used to ask for vehicle information and then return them: make,model,year,mileage
#this function doesn't take arguement(s)
#no need to handle exceptions of make input and model input
#handle exceptions to make sure year is a 4-digit positive number between 1900 and 2020
#handle exceptions to make sure mileage is a positive number
def get_vehicle_input():
    vmake = input("Enter Vehicle Make: ")
    vmodel = input("Enter Vehicle Model: ")
    err_flg = True
    while err_flg:
        try:
            vyear = int(input('Enter the year of manufacture between 1900 and 2020: '))
            if ((len(str(vyear)) < 4) or not(1900 <= vyear <= 2020)):
                raise ValueError
        except ValueError:
            print('\nYear of Manufacture should be 4-digit and between 1900 and 2020!')
        else:
            err_flg = False
    err_flg = True
    while err_flg:
        try:
            vmileage = int(input('Enter the mileage: '))
            if vmileage < 1: 
                raise ValueError            
        except ValueError:
            print('\nEnter an integer value for mileage!')
        else:
            err_flg = False
    return (vmake, vmodel, vyear, vmileage)
```

```python
#complete this read_file_data() function to prepare for the main menu application
#this function is used to read information from an existing databse/binary file, which is empdata.dat in the final exam
#this function doesn't take argument(s)
#this function returns a list containing all employee objects stored in the database; each employ object is an element in the list; the list will also be used and updated later
#need to handle exceptions if file doesn't exist
import pickle
def read_file_data():
    my_list = []
    try:
        fb = open('empdata.dat', 'rb')
    except FileNotFoundError:
        return my_list
    else:
        try:
            while(True):
                my_list.append(pickle.load(fb))
        except EOFError:
            fb.close()
            return my_list
```

```python
#complete option 5 related code here
#the function is used to show some basic statistics. They are: 
#(1) the number of employees stored in the database
#(2) the highest weekly compensation
#(3) the mean weekly compensation
#(4) the number of employees who have a vehicle with over 100,000 mileage
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#no need to handle exceptions

def run_option5(my_lst):
    compensation_list=[]
    veh_miles_emp_high_count = 0
    vehicle_miles_list=[]
    print("\n==========================================================================")
    print("Below is the Statistics of all employees in this database")
    print("==========================================================================")
    print("There are {} employees stored in the database.".format(len(emp_list)))
    for empl in my_lst:
        compensation_list.append(float(empl.compute_compensation()))
        vehicle_miles_list.append(float(empl.get_vehicle().get_mileage()))
    max_compensation = max(compensation_list)
    mean_compensation = sum(compensation_list)/len(compensation_list)
    print("The highest weekly compensation is ${:0.2f}".format(max_compensation))
    print("The mean weekly compensation is ${:0.2f}".format(mean_compensation))
    
    for veh in vehicle_miles_list:
        if veh > 100000:
            veh_miles_emp_high_count += 1
    print("The number of employees who have a vehicle with over 100,000 mileage is/are: {0}".format(veh_miles_emp_high_count))
```
